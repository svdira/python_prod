```python
import sqlite3
import os

def list_files_in_directory(directory_path):
    files = []
    try:
        # Get all entries (files and directories) in the specified path
        all_entries = os.listdir(directory_path)
        
        # Filter out directories and keep only files
        for entry in all_entries:
            full_path = os.path.join(directory_path, entry)
            if os.path.isfile(full_path):
                files.append(entry)
    except FileNotFoundError:
        print(f"Error: Directory not found at '{directory_path}'")
    except Exception as e:
        print(f"An error occurred: {e}")
    return files
```

```python
files = list_files_in_directory('media/wiki_media')
```

```python
conn = sqlite3.connect('db.sqlite3')

imgs_ok = []
```

```python
cursor = conn.cursor()
```

```python
cursor.execute("SELECT * FROM times_timesmedia;")
rows = cursor.fetchall()
for r in rows:
    imgs_ok.append(r[3])

cursor.execute("SELECT * FROM times_moviemedia;")
rows = cursor.fetchall()
for r in rows:
    imgs_ok.append(r[2])

cursor.execute("SELECT * FROM times_equipo;")
rows = cursor.fetchall()
for r in rows:
    imgs_ok.append(r[3])

cursor.execute("SELECT * FROM times_dirabook;")
rows = cursor.fetchall()
for r in rows:
    if len(r[7])>2:
        imgs_ok.append(r[7])
```
